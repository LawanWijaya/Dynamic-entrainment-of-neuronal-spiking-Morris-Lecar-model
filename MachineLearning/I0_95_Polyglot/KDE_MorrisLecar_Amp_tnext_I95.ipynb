```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FilteredData_MorrisLecar_I0_95.csv")
```

```python
plt.rcParams.update({
    'text.usetex': True,
    'font.family': "serif",
    'font.serif': ['Computer Modern Roman'],
    'axes.formatter.use_mathtext': False,
    'text.latex.preamble': r'\usepackage{amsmath}',
    'font.size': 28,
    'axes.titlesize': 28,
    'axes.labelsize': 28,
    'xtick.labelsize': 26,
    'ytick.labelsize': 26,
    'legend.fontsize': 26
})

single_color = "#1f77b4"

```

```python
# --- Plot Amp vs tnext ---
fig, ax = plt.subplots(figsize=(8,6))

sns.scatterplot(
    data=df, x="Amp", y="tnextmax",
    s=20, color=single_color, ax=ax
)

# Add 2D KDE contour
sns.kdeplot(
    data=df, x="Amp", y="tnextmax",
    levels=5, color=single_color, ax=ax
)

ax.set_xlabel(r"$Amp$")
ax.set_ylabel(r"$t_{next}$")
plt.tight_layout()
plt.savefig("Amp_vs_tnext.png", bbox_inches="tight")
plt.show()
```

```python
# --- Plot tnext vs Amp ---
fig, ax = plt.subplots(figsize=(8,6))

sns.scatterplot(
    data=df, x="tnextmax", y="Amp",
    s=20, color=single_color, ax=ax
)

sns.kdeplot(
    data=df, x="tnextmax", y="Amp",
    levels=5, color=single_color, ax=ax
)

ax.set_xlabel(r"$t_{next}$")
ax.set_ylabel(r"$Amp$")
plt.tight_layout()
plt.savefig("tnext_vs_Amp.png", bbox_inches="tight")
plt.show()
```

```python
# --- Plot L vs tnext ---
fig, ax = plt.subplots(figsize=(8,6))

sns.scatterplot(
    data=df, x="L", y="tnextmax",
    s=20, color=single_color, ax=ax
)

sns.kdeplot(
    data=df, x="L", y="tnextmax",
    levels=5, color=single_color, ax=ax
)

ax.set_xlabel(r"$L$")
ax.set_ylabel(r"$t_{next}$")
plt.tight_layout()
plt.savefig("L_vs_tnext.png", bbox_inches="tight")
plt.show()
```
