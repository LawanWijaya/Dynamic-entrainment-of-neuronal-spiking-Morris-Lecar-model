```python
import tensorflow.keras.backend as K
from tensorflow.keras.models import load_model

# re-define your custom activation exactly as before
def radbas(x):
    return K.exp(-K.square(x))

# load the .keras file
model4_matlab = load_model('model4_matlab.keras',
                         custom_objects={'radbas': radbas})

```

```python
import pandas as pd
```

```python
new_input = pd.DataFrame({
    'Amp': [30],
    'L': [20],
    'L1':[17],
    # 'p1':[40.02853],
    'L2':[92],
    'p2':[-27.5424] 
})
prediction_new = model4_matlab.predict(new_input.values)
print("4 Layer NN - Prediction with 5 inputs - tnext: {:.2f}".format(prediction_new[0][0]))
```
